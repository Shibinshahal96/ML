```python
swe1=swe.copy()
swe1
```

```python
x=swe1[["Annual Income (k$)","Spending Score (1-100)"]]
x.head()
wcss=[]
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python

wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(x)
    wcss.append(km.inertia_)
# import matplotlib.pyplot as plt
plt.plot(wcss,marker="o")
plt.xlabel("x values")
plt.ylabel("wcss")
plt.show
```
